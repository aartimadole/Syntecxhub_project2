```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset safely
df = pd.read_csv(r"E:\Sytecxhub_project2\time_series\Sample - Superstore.csv", encoding="latin1")
df
```

### Histogram


```python
plt.figure()
df["Sales"].hist(bins=30)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("sales_histogram.png")
plt.show()
plt.close()
```

### Boxplot by Region


```python

plt.figure()
df.boxplot(column="Sales", by="Region")
plt.title("Sales Distribution by Region")
plt.suptitle("")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.tight_layout()
plt.savefig("sales_boxplot_region.png")
plt.show()
plt.close()
```

### Outlier Detection (IQR)


```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df["Sales"] < Q1 - 1.5*IQR) |
              (df["Sales"] > Q3 + 1.5*IQR)]

with open("project2.2_summary.txt", "w") as f:
    f.write(f"Number of outliers detected: {len(outliers)}\n")
    f.write("Sales distribution appears right-skewed.\n")

print("Project 2.2 Completed âœ…")
```
